2項分布

$$
B(x) :=
\begin{pmatrix}
n \\
x \\
\end{pmatrix}
p^x(1-p)^(n-x)
$$

```python
from scipy.stats import binom
```

```python
n = 10000
p = 0.55
rv = binom(n, p)

mu = rv.expect()
print(mu)
```

```python
var = rv.var()
var
```

```python
rv.rvs(5)
```
